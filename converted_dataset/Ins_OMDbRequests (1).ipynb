```python
movie_list=df['imdbId'].tolist()
```

```python
full_plot=[]
for movie in movie_list:
    try:
        res= omdb.request(i='tt00'+str(movie), plot='full',apikey='7329004f')
        data=res.json()
        full_plot.append({"Plot":data['Plot'], "Title":data['Title'],"Year":data['Year']})
    except:
        full_plot
```
