```python
# Import Dependencies
import glob
import pandas as pd
import os
```

```python
# Use glob to load all CSVs
path = r"movies_data_with_full_info"
all_files = glob.glob(os.path.join(path, "*.csv"))
records_df = [pd.read_csv(file) for file in all_files]
# print(records_df)
```

```python
full_movie_data = pd.concat(records_df, ignore_index=True)
```

```python
full_movie_data
```

```python
full_movie_data.to_csv("full_movie_data.csv", index=False)
```
