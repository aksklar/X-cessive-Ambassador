```python
import omdb
import pandas as pd
import csv
from pprint import pprint
```

```python
df=pd.read_csv('data/movies_100.csv')
df.head()
```

```python
len(df)
```

```python
movie_list=df['imdbId'].tolist()
```

```python
movie_counter = 1
full_plot=[]
for movie in movie_list:
    try:
        res= omdb.request(i='tt0'+str(movie), plot='full',apikey='f43f79fb')
        data=res.json()
        print(f"Processing Record {movie_counter} | {data['Title']}")
        full_plot.append({"Title":data['Title'], "Year":data['Year'], "Plot":data['Plot'], "Rated":data['Rated'], "Genre(s)":data['Genre']})
        movie_counter += 1
    except:
        print(f"Data does not exist for this movie: {data['Title']}")
        continue
print("Done Retrieving Records")
```

```python
full_plot
```

```python
len(full_plot)
```

```python
movie_df=pd.DataFrame(full_plot)
```

```python
movie_df=movie_df[["Title", "Year", "Rated", "Genre(s)", "Plot"]]
```
