```python
# Import Dependencies
import glob
import pandas as pd
import os
```

```python
# Use glob to load all CSVs
path = r"movie_data_with_plot"
all_files = glob.glob(os.path.join(path, "*.csv"))
records_df = [pd.read_csv(file) for file in all_files]
# print(records_df)
```

```python
full_movie_data_mega= pd.concat(records_df, ignore_index=True)
```
